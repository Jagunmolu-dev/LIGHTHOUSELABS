# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
# Get request from api
url = "https://api.foursquare.com/v3/places/search?ll=52.668746%2C-8.627809&radius=1000&fields=categories%2Crating%2Cgeocodes%2Ccategories%2Clocation%2Cname%2Cfsq_id%2Cstats%2Cprice&limit=20"

headers = {
    "accept": "application/json",
    "Authorization": "fsq3XN1vgbF0VBUV5pSmRqSCkjKpIn0n/S2PEtkckuL7ENo="
}

response = requests.get(url, headers=headers)

data =response.json()
df_1 = pd.DataFrame(data['results'])

# Unpack the json and extract the required columns
df_2 = df_1.copy()
df_2 = df_2[['name', 'rating']]


geo_df = df_1['geocodes']
geo_df = pd.json_normalize(geo_df)

cat_df = df_1['categories']
cat_df = pd.json_normalize(cat_df)
cat_df = pd.json_normalize(cat_df[0])
cat_df = cat_df.dropna()

cat_df = pd.DataFrame(cat_df['name'])
cat_df.rename(columns = {'name': 'category'}, inplace = True) 

loc_df = df_1['location']
loc_df = pd.json_normalize(loc_df)

loc_df = loc_df.drop(['locality', 'region', 'cross_street', 'postcode', 'formatted_address', 'country'], axis = 1)
loc_df = loc_df.dropna()

stats_df = df_1['stats']
stats_df = pd.json_normalize(stats_df)
stats_df = stats_df[['total_ratings']]
stats_df = stats_df.dropna()

# combine the different dataframes
fsq_df = pd.concat([df_2, geo_df, cat_df, loc_df, stats_df], axis = 1)
fsq_df= fsq_df[['name','address', 'category', 'rating', 'main.latitude', 'main.longitude']]

```
