# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
# imports
import os
import requests
from IPython.display import JSON
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
# get response from the api
url = "https://api.citybik.es/v2/networks"
response = requests.get(url)
response
```

```python
# pass response as JSON
data = response.json()
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
data['networks'][150]['id']
```

```python
#create a dataframe from the network dictionary
networks = data['networks']
networks = pd.DataFrame.from_dict(networks)

```

```python
#Loop through the network to get all the network ids
for network in networks:
    network_id = networks['id']
print("Network ID: ", network_id)
```

```python
#get data for limerick network_id and pass it as JSON
network_id = 'limerick'
network_json = requests.get("http://api.citybik.es/v2/networks/"+network_id).json()
#JSON(network_json)
```

Put your parsed results into a DataFrame.

```python
# get data about the stations
df = pd.DataFrame.from_dict(network_json)
df = df['network']['stations']
df = pd.DataFrame.from_dict(df)
df.head()
```

```python
cols = ['extra', 'id']
df.drop(cols, axis = 'columns', inplace=True)
```

```python
citybike = df.copy()
citybike.head()
```

```python
# save dataframe as csv
citybike.to_csv('citybike.csv')
```
